### **Metrics Evaluation Pada Model Regresi:**

Penentuan Metrics Evaluation pada Model yang dibuat hendaknya disesuaikan dengan model yang akan dibangun, untuk penentuan prediksi harga rumah berdasarkan dataset ini metrics evaluation yang tepat dan umum digunakan untuk menghitung tingkat keakuratan model regresi adalah sebagai berikut:

1. 	**Mean Absolute Error (MAE):**<br>
Deskripsi: MAE mengukur rata-rata dari selisih absolut antara prediksi dan nilai sebenarnya.<br>
Keuntungan: Mudah diinterpretasikan karena memberikan ukuran rata-rata kesalahan dalam unit yang sama dengan variabel target.<br>
Catatan: Tidak peka terhadap outlier.

2. 	**Mean Squared Error (MSE):**<br>
Deskripsi: MSE mengukur rata-rata dari kuadrat selisih antara prediksi dan nilai sebenarnya.<br>
Keuntungan: Sensitif terhadap kesalahan besar karena memperkuat pengaruh outlier.<br>
Catatan: Nilai MSE lebih sulit diinterpretasikan karena unitnya adalah kuadrat dari unit target.

3. 	**Root Mean Squared Error (RMSE):**<br>
Deskripsi: RMSE adalah akar kuadrat dari MSE. Ini memberikan ukuran kesalahan dalam unit yang sama dengan variabel target.<br>
Keuntungan: Lebih mudah diinterpretasikan dibandingkan MSE karena memiliki unit yang sama dengan variabel target.<br>
Catatan: Seperti MSE, RMSE juga dapat dipengaruhi oleh outlier.

4. 	**R-squared (R2):**<br>
Deskripsi: R2 mengukur sejauh mana variasi dalam variabel target dapat dijelaskan oleh model.<br>
Keuntungan: Memberikan informasi tentang seberapa baik model Anda memfitting data.<br>
Catatan: Nilai R2 berkisar antara 0 dan 1, dan semakin mendekati 1, semakin baik modelnya. Namun, R2 dapat memberikan hasil yang bias jika digunakan pada model yang kompleks.

### **Metrics Evaluation yang Kami Gunakan:**

1. 	**Median Absolute Error (MedAE):**<br>
Deskripsi: MedAE mengukur median dari selisih absolut antara prediksi dan nilai sebenarnya.<br>
Keuntungan: Lebih tahan terhadap outlier dibandingkan MAE.<br>
Catatan: Lebih sesuai jika data memiliki outlier yang signifikan.

```python
def find_outlier(df, feature):
    print('Outlier ' + feature)
    q1 = df[feature].quantile(0.25)
    q3 = df[feature].quantile(0.75)
    iqr = q3 - q1
    limit = iqr*1.5
    print(f'IQR: {iqr}')

    limit_bawah = q1 - limit
    limit_atas = q3 + limit
    print(f'limit_bawah: {limit_bawah}')
    print(f'limit_atas: {limit_atas}')
    print('_________________________')
```

### **1. Encoding & Transformasi pada Kolom Category**

Proses pertama yang akan dilakukan pada tahapan **Feature Engineering** ini adalah **Encoding** pada kolom-kolom categorical, yang pada konteks data saya adalah kolom `ocean_proximity` dan `house_age_category`.<br>
Untuk metode Encoding yang saya gunakan pada kolom `ocean proximity` adalah metode **OHE/One-Hot Encoding**, dan untuk ini saya memanfaatkan fungsi pandas.get_dummies(). Hal ini dipilih karena kolom categorical ini hanya memiliki unique value yang terbatas dan sedikit (5 unique value) dan tidak memiliki urutan/tingkatan pada masing-masing value nya.

Untuk metode Encoding pada kolom `house_age_category` saya menggunakan metode **OrdinalEncoder** dikarenakan value unique pada kolom ini bisa dikatakan memiliki urutan/tingkatan pada masing-masing value nya.

Untuk menghindari multicollinearity, yang dapat mempengaruhi beberapa model pembelajaran mesin, biasanya kita bisa menghapus salah satu kolom pada hasil encoding.<br> 

Menurut buku "Introduction to Linear Regression Analysis" oleh Montgomery, Peck, and Vining (1982), Multicollinearity adalah kondisi yang terjadi ketika dua atau lebih variabel dalam model memiliki korelasi yang tinggi satu sama lain. Dalam konteks OHE, jika kita menyimpan semua kolom yang dihasilkan dari encoding (menggunakan 1 untuk menunjukkan keanggotaan kategori dan 0 untuk yang lainnya), kolom-kolom tersebut akan saling terkait secara linear. Ini disebabkan oleh sifat dasar bahwa jika semua kecuali satu kolom adalah 0, maka kolom yang tersisa harus bernilai 1. Dengan kata lain, satu kolom dapat dihasilkan dari kolom-kolom yang lain.<br>

Untuk pemilihan kolom mana yang dihapus biasanya bebas, akan tetapi pada konteks kami, kami mempertimbangkan untuk menghapus kolom `ocean_proximity_ISLAND` dengan pertimbangan ketidak seimbangan data dimana data yang mengandung kategori tersebut hanya 2 data.

Detail alasan kami memanfaatkan metode ini adalah sebagai berikut:<br>
- Robust Scaler cocok untuk data yang tidak terdistribusi normal.

- Penskalaan rentang interquartile jadi dapat membantu dalam memahami sebaran data dan mengurangi dampak dari nilai ekstrem outlier.
- Robust Scaler tetap membantu pemeliharaan nilai median dan quartile.

### **2. Model Benchmarking**

Tahapan pertama pada proses modeling ini adalah Benchmarking, yaitu perbandingan dan evaluasi kinerja berbagai model atau metode pada tugas tertentu yang diinginkan.<br>

Pada Tahapan ini, kami mencoba membandingkan kinerja model regresi dengan melakukan proses training menggunakan data train dan melakukan pengujian menggunakan data test untuk menemukan dan menentukan model terbaik yang akan digunakan untuk memprediksi harga rumah. Berikut detail tahapan yang kami lakukan:

- **Menentukan tujuan Benchmarking**<br>
Pada tahap ini kami menentukan tujuan benchmarking untuk melakukan tugas regresi, ooleh karena itu kami menentukan berbagai model atau metode yang umum digunakan pada tugas regresi, seperti: **LinearRegresi, Lasso, Ridge, KNN-Regressor, Decision Tree, Random Forest, AdaBoostR, GradientBoostR, XGBRegressor, dan SVR**.

- **Menentukan Metrik Evaluasi**<br>
Pada tahap ini, kami menentukan metrik evaluasi yang akan digunakan untuk scoring model, pada konteks kami, kami menentukan untuk menggunakan metode scoring `Med-AE (Median-Absolute Error)`, dengan alasan `Median-Absolute Error (Med-AE)` digunakan untuk mengukur seberapa dekat prediksi model dengan nilai sebenarnya, dengan fokus pada median dari selisih absolut antara prediksi dan nilai sebenarnya. Med-AE sangat berguna ketika data mengandung pencilan atau nilai yang signifikan, karena median kurang sensitif terhadap nilai ekstrem daripada rata-rata (mean). Karena pada konteks data kami, data yang digunakan memiliki outlier yang cukup banyak dan beberapa nilai memiliki nilai yang cukup signifikan dengan nilai rata-rata data. Penting untuk diingat bahwa Med-AE lebih tahan terhadap pencilan (outliers) daripada metrik seperti Mean Squared Error (MSE), karena Med-AE tidak menghitung kuadrat selisih. Oleh karena itu, Med-AE dapat memberikan gambaran yang lebih stabil tentang kesalahan prediksi model jika terdapat banyak nilai yang jauh dari rata-rata. Semakin kecil nilai Med-AE, semakin baik kinerja model, karena menunjukkan bahwa prediksi lebih dekat dengan nilai sebenarnya.

- **Analisa, Optimasi, dan Pemilihan Model Terbaik**


```python
# Algoritma ML yang digunakan
lr = LinearRegression()
lasso = Lasso(random_state = 27)
ridge = Ridge(random_state = 27)
knn = KNeighborsRegressor()
dt = DecisionTreeRegressor(random_state=27)
rf = RandomForestRegressor(random_state=27)
ada = AdaBoostRegressor(random_state = 27)
gbr = GradientBoostingRegressor(random_state = 27)
xgb = XGBRegressor(random_state=27)
svm = SVR()
```

```python
models = [lr, knn, dt, rf, xgb, lasso, ridge, ada, gbr, svm]

estimator = Pipeline([
        ('preprocessing', transformer),
        ('model', None)
        ])

param_space = {
    'model' : models,
    'preprocessing__Scalling' : [RobustScaler()]
}

crossval = KFold(n_splits=5, shuffle=True, random_state=27)

# Hyperparameter tuning
gridsearch_scaler = GridSearchCV(
    estimator = estimator,
    param_grid = param_space,
    cv = crossval, 
    scoring = ['neg_median_absolute_error'], 
    n_jobs = -1,
    refit='neg_median_absolute_error',
    return_train_score=True
)
```

### **4. Hyperparameter Tunning**

Tahapan selanjutnya adalah melakukan HyperParameter Tunning pada model-model yang telah dipilih sebelumnya, yaitu: `GradientBoost-R`. Tujuan dilakukan tahapan ini adalah untuk mencari kombinasi nilai parameter paling optimal untuk hyperparameter suatu model machine learning guna meningkatkan kinerja model tersebut. Untuk menentukan parameter terbaik ini, kami memanfaatkan pendekatan metode `gridSearch` untuk mencari kombinasi nilai parameter terbaik di masing-masing model, metode `gridSearch` ini sendiri adalah teknik pencarian sistematis melalui berbagai kombinasi hyperparameter untuk menemukan kombinasi yang memberikan kinerja terbaik pada model machine learning. Teknik ini digunakan untuk melakukan hyperparameter tuning dengan cara menguji setiap kombinasi hyperparameter yang telah ditentukan pada suatu "grid" yang telah ditentukan sebelumnya.

#### **Hyperparameter Model GradientBoost-R**

Pada model `GradientBoost-R` penentuan kombinasi parameter yang digunakan adalah parameter-parameter yang umum digunakan pada model ini dengan pertimbangan pada fungsi masing-masing parameter pada model `GradientBoost-R`, yaitu:

- learning rate: Laju pembelajaran, yaitu ukuran langkah yang diambil model saat menyesuaikan parameternya
- n_estimators: Jumlah pohon dalam model
- max depth: Kedalaman maksimum pohon tunggal
- alpha: Parameter alpha mengontrol regularisasi L1 (regularisasi Lasso) pada bobot pohon. Regularisasi L1 menambahkan jumlah absolut dari bobot sebagai bagian dari fungsi tujuan yang dioptimalkan, membantu mencegah overfitting dan meningkatkan generalisasi model.

Untuk rentang nilai yang digunakan pada masing-masing parameter adalah rentang nilai-nilai yang umum digunakan pada tuning parameter. Dengan pertimbangan sebagai berikut: 

1. learning rate : Kami ingin mencari kecepatan model dalam mempelajari data yang tepat untuk model prediksi.
2. n_estimators: Dikarenakan pada model benchmarking ditemukan overfitting pada data, maka n_estimators yang digunakan tidak terlalu tinggi.
3. max depth: Penggunaan nilai max depth dipilih dengan nilai-nilai yang cukup rendah dengan tujuan tidak menambah overfitting pada model.
4. alpha: Pemilihan parameter ini didasarkan pada keinginan kami untuk mengontrol regulasi Lasso pada bobot pohon.

Source: "A Comprehensive Survey on Extreme Gradient Boosting for House Price Prediction" oleh Wang dkk. (2023)

```python
estimator = Pipeline([
        ('preprocessing', transformer),
        ('model', None)
        ])

param_space_gb = {
    'model' : [GradientBoostingRegressor(random_state = 27)],
    'model__learning_rate': [0.01,0.1],
    'model__n_estimators' : [100,300],
    'model__max_depth': [3,5,7],
    'model__alpha':[0.1, 0.9],
    'preprocessing__Scalling' : [RobustScaler()]
}

crossval = KFold(n_splits=5, shuffle=True, random_state=27)

# Hyperparameter tuning
best_gb = GridSearchCV(
    estimator = estimator,
    param_grid = param_space_gb,
    cv = crossval, 
    scoring = 'neg_median_absolute_error',
    n_jobs = -1,
    verbose=2
)
```

    - Estimasi harga prediksi yang lebih akurat daripada prediksi secara manual, karena adanya pertimbangan-pertimbangan atribut yang berpengaruh pada tinggi rendahnya prediksi harga rumah.
    - Saving time dalam memprediksi harga rumah, jadi tidak perlu melakukan survey langsung ke lokasi rumah, asalkan data yang diberikan terkait kondisi rumah telah jelas dan detail.
    - Prediksi manual mungkin kurang fleksibel ketika dihadapkan pada perubahan kondisi pasar atau pola yang kompleks. Model machine learning dapat diubah dan disesuaikan dengan cepat untuk mencerminkan perubahan dalam data atau tren pasar.
    - Machine learning memiliki kemampuan untuk menangkap pola yang kompleks dan hubungan non-linear dalam data, yang mungkin sulit atau tidak dapat diidentifikasi dengan mudah oleh analisis manual.
